```python
import matplotlib.pyplot as plt
from curve import image_to_hilbert_curve
```

```python
image_tomato = plt.imread("./images/pomodoro.png")
image_stairs = plt.imread("./images/paolo-bendandi.jpg")/255
```

```python
curve_tomato = image_to_hilbert_curve(image_tomato)
```

```python
curve_stairs = image_to_hilbert_curve(image_stairs)
```
